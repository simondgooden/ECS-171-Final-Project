```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Weather-related disease prediction.csv')

numeric_cols = [ 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']
correlation = df[numeric_cols].corr()

plt.figure(figsize=(6, 5))
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45)
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title('Correlation Between Numeric Features')
plt.tight_layout()
plt.show()

symptom_1 = 'high_fever'
symptom_2 = 'joint_pain'
symptom_combination = pd.crosstab(df[symptom_1], df[symptom_2])
symptom_combination.plot(kind='bar', stacked=True, figsize=(6, 6), color=['lightblue', 'teal'])
plt.title(f'Combination of {symptom_1} and {symptom_2}')
plt.xlabel(symptom_1)
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

high_fever_data = df[df['high_fever'] == 1]
plt.figure(figsize=(8, 6))
plt.scatter(high_fever_data['Temperature (C)'], [1] * len(high_fever_data), alpha=0.6, color='teal')
plt.title('Temperature vs High Fever (Only High Fever = 1)')
plt.xlabel('Temperature (°C)')
plt.yticks([1], ['High Fever (1)'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
