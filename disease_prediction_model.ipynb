```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Weather-related disease prediction.csv')

continuous_vars = ['Age', 'Temperature (C)', 'Wind Speed (km/h)']

#boxplot to visualize our continous vars
plt.figure(figsize=(8, 4))
sns.boxplot(data=df[continuous_vars])
plt.title("Plotting Continuous Vars")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#prognosis bar chart
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='prognosis', order=df['prognosis'].value_counts().index)
plt.title('Number of Patients per Prognosis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
start_col = 'nausea'
end_col = 'prognosis'

# Create a list of column names between 'nausea' and 'prognosis' (inclusive)
cols = df.columns[df.columns.get_loc(start_col):df.columns.get_loc(end_col) + 1]

# Filter the relevant columns for binary data and prognosis
binary_data = df[cols]

# Calculate correlation between binary symptoms and prognosis
correlation_matrix = binary_data.corr()

# Plot a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation between Symptoms and Prognosis')
plt.tight_layout()
plt.show()
```
